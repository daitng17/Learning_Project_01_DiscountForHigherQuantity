## <center> Intro to Dataframes

```python
import pandas as pd
```

```python
scores = {"name":['Ray','Japhy','Zosa'],
          "city":['San Francisco','San Francisco','Denver'],
          "score":[75,92,94]
         }
```

```python
df = pd.DataFrame(scores)
```

```python
df
```

```python
df['score']
```

```python
df['name_city'] = df['name'] + '_' + df['city']
```

```python
df[df['score']>90]
```

```python
# Create a dataframe consists of places that I want to visit

import pandas as pd

toGoList = {
    'city' : ['Osaka', 'Vung Tau', 'Ha Noi', 'Hokkaido', 'Seoul', 'London', 'Paris'],
    'country' : ['Japan', 'Vietnam', 'Vietnam', 'Japan', 'Korea', 'England', 'France'],
    'year' : [2019, 2022, 2022, 2022, 2023, 2024, 2024],
    'visited' : ['Yes', 'No', 'No', 'No', 'No', 'No', 'No']
}

df = pd.DataFrame(toGoList)
df
```

```python
df.country
```

```python
df[df.country == 'Vietnam']
```

```python
df['destination'] = df['city'] + ', ' + df['country']
```
